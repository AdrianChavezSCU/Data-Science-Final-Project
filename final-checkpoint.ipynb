```python
#Here we decided to trim down the dataset and only work with variables that we were interested and thought might have
#a correlation with police killings. 


new_df = df[ ['age', 'raceethnicity', 'armed', 'h_income','p_income', 'pov', 'urate','college'] ]

#We immediately took out any rows that did not have any inputs in them for age, raceethnicity, and poverty rates. 
new_df = new_df[new_df.age != 'Unknown']
new_df = new_df[new_df.raceethnicity != 'Unknown']
new_df = new_df[new_df.pov != '-']

#Some of the columns from the dataset were entered as objects as opposed to floats or ints so we had to change that. 
new_df['age'] = pd.to_numeric(new_df.age)
new_df['pov'] = pd.to_numeric(new_df.pov)
new_df['p_income'] = pd.to_numeric(new_df.p_income)
```

```python
#We decided to scale the variables so that we would be able to look at their boxplots all at once as well as have 
#everything on the same scale from 0 to 1. 
def scale_variables(new_df):
    # Household income in thousands of dollars 
    new_df.h_income = new_df.h_income / (100000)
    # Personal income in thousands of dollars 
    new_df.p_income = new_df.p_income / (100000)
    # Poverty rate in percentage from 0 to 1 
    new_df.pov = new_df.pov / (100)
    
scale_variables(new_df)
```

```python
#Here we ran our regression model using age as our outcome variable and r_Hispanic and r_Black as our predictor dummy variables.
model_age = smf.ols(formula = 'age ~ r_Hispanic + r_Black', data = new_df).fit()
#print new_df.loc[new_df['r_Hispanic'] == 1].age.mean()
model_age.summary()
```

```python
#As we can see above the y-intercept above is corresponding to the average age of White people who are killed by police. 
#The coefficient values of -8.71 for r_Hispanic means that on average, people who are killed by 
#police that are Hispanic have an average age of 40.4936 - 8.9027 = 31.5729
#Likewise, the average age of people who are black and killed by police is 40.4936 - 6.7192 = 33.7746

figure = sm.qqplot(model_age.resid, line = 's')
```

```python
train_df = new_df.sample(frac = .6, random_state = 0).sort()
test_df = new_df.drop(train_df.index)

train_X = train_df[ ['age','h_income','p_income','pov','urate','college'] ]
train_y = train_df['raceethnicity']

model = neighbors.KNeighborsClassifier(n_neighbors = 5, weights = 'uniform')
model.fit(train_X, train_y)
print 'train = ', model.score(train_X, train_y)

test_X = test_df[ ['age','h_income','p_income','pov','urate','college'] ]
test_y = test_df['raceethnicity']

print 'test  = ', model.score(test_X, test_y)
```

```python
k = range(1, len(train_df))

score_df = pd.DataFrame({'k': k})

models = score_df.\
    apply(lambda row: neighbors.KNeighborsClassifier(n_neighbors = row['k']).fit(train_X, train_y), axis = 1)

score_df['TrainScore'] = models.apply(lambda model: model.score(train_X, train_y))
score_df['TestScore'] = models.apply(lambda model: model.score(test_X, test_y))
```

```python
k_cv = 10 # 5-fold CV
k_nn = range(1, len(train_df) * (k_cv - 1) / k_cv) # k-NN

gs = grid_search.GridSearchCV(
    estimator = neighbors.KNeighborsClassifier(),
    param_grid = {'n_neighbors': k_nn},
    cv = cross_validation.KFold(len(train_df), n_folds = k_cv)
)

gs.fit(train_X, train_y)

score_df = pd.DataFrame({'k': [score.parameters['n_neighbors'] for score in gs.grid_scores_],
    'Score': [score.mean_validation_score for score in gs.grid_scores_]})

score_df
```
